### Data Cleaning

```python
## Missing values

df.isnull().sum()
```

```python
df_num = df_num.dropna()
```

```python
df_num.isnull().sum()
```

```python
x = df_num.drop(["Id","SalePrice"],axis = 1)
y = df_num["SalePrice"]
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size = 0.7, test_size = 0.3, random_state = 100)
LR = LinearRegression()
LR.fit(xtrain,ytrain)
train_y_pred = LR.predict(xtrain)
test_y_pred = LR.predict(xtest)
print("RMSE:",np.sqrt(mean_squared_error(ytest,test_y_pred)))
```

```python
test_df = pd.read_csv("test.csv")
test_df_num = test_df.select_dtypes(exclude = "object")
test_df_num.shape
test_df_num.isnull().sum()
test_df_num["LotFrontage"].fillna(test_df_num["LotFrontage"].mean(),inplace = True)
test_df_num["MasVnrArea"].fillna(test_df_num["MasVnrArea"].mean(),inplace = True)
test_df_num["BsmtFinSF1"].fillna(test_df_num["BsmtFinSF1"].mean(),inplace = True)
test_df_num["BsmtFinSF2"].fillna(test_df_num["BsmtFinSF2"].mean(),inplace = True)
test_df_num["BsmtUnfSF"].fillna(test_df_num["BsmtUnfSF"].mean(),inplace = True)
test_df_num["TotalBsmtSF"].fillna(test_df_num["TotalBsmtSF"].mean(),inplace = True)
test_df_num["BsmtFullBath"].fillna(test_df_num["BsmtFullBath"].mean(),inplace = True)
test_df_num["BsmtHalfBath"].fillna(test_df_num["BsmtHalfBath"].mean(),inplace = True)
test_df_num["GarageYrBlt"].fillna(test_df_num["GarageYrBlt"].mean(),inplace = True)
test_df_num["GarageCars"].fillna(test_df_num["GarageCars"].mean(),inplace = True)
test_df_num["GarageArea"].fillna(test_df_num["GarageArea"].mean(),inplace = True)
```
